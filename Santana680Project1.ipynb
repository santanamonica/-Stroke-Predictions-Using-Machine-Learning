```python
# Import library
import pandas as pd

# Remove warnings
import warnings
warnings.filterwarnings('ignore')

# Import dataset
stroke_df = pd.read_csv('Downloads/stroke_data.csv')
stroke_df
```

```python
# Checking for missing data
# BMI column has 201 NaN's

stroke_df.isnull().sum()
```

```python
# Import library
import numpy as np

# Drop the NaN's in the BMI column
stroke_df.dropna(inplace=True)
stroke_df
```

```python
# Checking the bmi column
# Clear of NaN's

stroke_df.isnull().sum()
```

```python
# Drop columns that I will not be using using drop() function.
# Dropping "id", "ever_married", "work_type", and "Residence_type"

stroke_data = stroke_df.drop(columns=['id', 'ever_married', 'work_type', 'Residence_type'], axis = 1)
stroke_data.head(15)
```

```python
# Check all data types

stroke_data.dtypes
```

```python
# Looking over at the types I want to convert age from a float type to an integer, I don't need to account for halves

stroke_data['age'] = stroke_data['age'].astype('int64')
```

```python
# Now the dataset is ready for modeling

stroke_data.head(15)
```

```python
# Convert categorical columns to dummy variables and preview the dataset to make sure it is converted

categorical = stroke_data.select_dtypes(include = ['object']).columns.tolist()
dummiesdf = pd.get_dummies(stroke_data, columns = categorical)
dummiesdf
```

```python
# Split the data into a training and test set, where the "stroke" column is the target for my models

# Import libraries
from sklearn.model_selection import train_test_split

x = dummiesdf.drop('stroke', axis=1)
y = dummiesdf['stroke']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
# Run and fit a logistic regression to the training set

# Import libraries
from sklearn.linear_model import LogisticRegression

# Variable to create the logistic regression model
logisticmodel = LogisticRegression()

# Train logistic regression
logisticmodel.fit(xtrain, ytrain)
```

```python
# Calculating the results for test data, lm at the end of the variables stands for logistic model
ytestpredlm = logisticmodel.predict(xtest)

# Import library
from sklearn import metrics

# Print the accuracy score for the logistic regression model
metrics.accuracy_score(ytest, ytestpredlm)*100
```

```python
# Import library
from sklearn.metrics import classification_report

# Print the classification report of the metrics
print(classification_report(ytest, ytestpredlm))
```

```python
# Trying a different model to see if there is improvement with the random forest classifier

# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier

# Variable to create the random forest model
rforestmodel = RandomForestClassifier()

# Train random forest classifier
rforestmodel.fit(xtrain, ytrain)
```

```python
# Import library
from sklearn import metrics

# Calculating the results for test data for random forest where the end of the variables has rf for random forest
ytestpredrf = rforestmodel.predict(xtest)

metrics.accuracy_score(ytest, ytestpredrf)*100
```

```python
# Printing a classification report of the results

from sklearn.metrics import classification_report

print(classification_report(ytest, ytestpredrf, zero_division=1))
```

```python
# Create a new variable for important features using the random forest classifier model

rforestimportance = rforestmodel.feature_importances_
rforestimportance
```

```python
# Create a variable for important features and the names of the columns

rforestnames = rforestmodel.feature_names_in_
rforestnames
```

```python
# Create a new dataframe for the important features to plot them as a bar graph

importance_df = pd.DataFrame({"Feature Importances" : rforestimportance, 
                               "Feature Names" : rforestnames})
importance_df
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the important features for the stroke dataset using a bar chart
sns.barplot(x='Feature Importances', y='Feature Names', data = importance_df)
plt.title('Feature Importances for Strokes')
plt.show()
```
